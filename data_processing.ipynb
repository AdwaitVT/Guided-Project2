```python
import pandas as pd
import datetime as datetime
import numpy as np

acoustic_features = pd.read_csv('acoustic_features.csv')
albums = pd.read_csv('albums.csv')
```

```python
albums
```

```python
import datetime

#drop everything from before 1960
acoustic_features['date'] = pd.to_datetime(acoustic_features['date'])
acoustic_features = acoustic_features[acoustic_features['date'].dt.year >= 1960]

acoustic_features.head()
```

```python
# take album info to merge later
album_info = acoustic_features[['album', 'artist', 'album_id', 'date']]
album_info = album_info.drop_duplicates()
album_info.head()
```

```python
#take just the data values
data_values = acoustic_features.drop(acoustic_features.columns[[0, 1, 3]], axis=1) 
data_values = data_values.drop(['album_id', 'date'], axis=1) 

data_values.head()
```

```python
#average values over the album
album_values = data_values.groupby(['album']).mean()
album_values.head()
```

```python
#merge averaged values with album info
album_data = pd.merge(album_info, album_values, on='album')
album_data = album_data.sort_values(by=['date'])
album_data
```

```python
#take album rankings
album_ranking = albums[['date','album', 'rank']]
album_ranking = album_ranking.dropna()
album_ranking['date'] = pd.to_datetime(album_ranking['date'])
album_ranking
```

```python
#merge values with chart ranking
albums = pd.merge(album_ranking, album_data, on="album")
albums = albums.rename(columns={"date_x": "chart_date", "date_y": "release_date"})
albums = albums.sort_values(by=['chart_date', 'rank'])
albums = albums.reset_index()
albums = albums.drop(['index'], axis=1)
albums
```

```python
#remove albums where chart date was before release date
albums = albums[albums['chart_date'] >= albums['release_date']]
albums = albums.reset_index()
albums = albums.drop(['index'], axis=1)
albums
```

```python
albums.head()
```

```python
yearly_avgs = albums.drop(['rank','key', 'mode', 'time_signature'], axis=1) 
yearly_avgs = yearly_avgs.groupby(albums.chart_date.dt.year).mean()
yearly_avgs = yearly_avgs.reset_index()
yearly_avgs.head()
```

```python
yearly_avgs.to_csv('yearly_avgs.csv', encoding='utf-8', index=False)
```

```python
import datetime
sixties = datetime.datetime(1960, 1, 1)
seventies = datetime.datetime(1970, 1, 1)
eighties = datetime.datetime(1980, 1, 1)
nineties = datetime.datetime(1990, 1, 1)
aughts = datetime.datetime(2000, 1, 1)
tens = datetime.datetime(2010, 1, 1)
print("60s:", len(albums[(albums['chart_date'] >= sixties) & (albums['chart_date'] < seventies)]))
print("70s:", len(albums[(albums['chart_date'] >= seventies) & (albums['chart_date'] < eighties)]))
print("80s:", len(albums[(albums['chart_date'] >= eighties) & (albums['chart_date'] < nineties)]))
print("90s:", len(albums[(albums['chart_date'] >= nineties) & (albums['chart_date'] < aughts)]))
print("00s:", len(albums[(albums['chart_date'] >= aughts) & (albums['chart_date'] < tens)]))
print("10s:", len(albums[(albums['chart_date'] >= tens)]))
```

```python
# store data by decade
d60 = albums[(albums['chart_date'] >= sixties) & (albums['chart_date'] < seventies)]
d70 = albums[(albums['chart_date'] >= seventies) & (albums['chart_date'] < eighties)]
d80 = albums[(albums['chart_date'] >= eighties) & (albums['chart_date'] < nineties)]
d90 = albums[(albums['chart_date'] >= nineties) & (albums['chart_date'] < aughts)]
d00 = albums[(albums['chart_date'] >= aughts) & (albums['chart_date'] < tens)]
d10 = albums[(albums['chart_date'] >= tens)]

avg60s = d60.drop(['rank','key', 'mode', 'time_signature'], axis=1) 
avg60s = avg60s.groupby(albums.artist).mean()
avg60s = avg60s.reset_index()

avg70s = d70.drop(['rank','key', 'mode', 'time_signature'], axis=1) 
avg70s = avg70s.groupby(albums.artist).mean()
avg70s = avg70s.reset_index()

avg80s = d80.drop(['rank','key', 'mode', 'time_signature'], axis=1) 
avg80s = avg80s.groupby(albums.artist).mean()
avg80s = avg80s.reset_index()

avg90s = d90.drop(['rank','key', 'mode', 'time_signature'], axis=1) 
avg90s = avg90s.groupby(albums.artist).mean()
avg90s = avg90s.reset_index()

avg00s = d00.drop(['rank','key', 'mode', 'time_signature'], axis=1) 
avg00s = avg00s.groupby(albums.artist).mean()
avg00s = avg00s.reset_index()

avg10s = d10.drop(['rank','key', 'mode', 'time_signature'], axis=1) 
avg10s = avg10s.groupby(albums.artist).mean()
avg10s = avg10s.reset_index()



artist60s = d60.artist.tolist()
artist70s = d70.artist.tolist()
artist80s = d80.artist.tolist()
artist90s = d90.artist.tolist()
artist00s = d00.artist.tolist()
artist10s = d10.artist.tolist()
```

```python
artistNames = []
totals = {}

artistNames = np.unique(artist60s)

for ii in range(len(artistNames)):
    #print(artistNames[ii])
    count = artist60s.count(artistNames[ii])
    totals[artistNames[ii]] = count

# sort
sortedTotals = dict(sorted(totals.items(), key=lambda item: item[1], reverse=True))

totalsDF = pd.DataFrame.from_dict(sortedTotals, orient='index')
totalsDF.head(10)
    
```
